## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
# Make copies of original pieces of data
t_api_df_copy = t_api_df.copy()
image_pred_copy = image_pred.copy()
tweet_arch_copy = tweet_arch.copy()
```

#### Just we can merge all copies of our data frames we have to make sure that tweet_id is str

```python
t_api_df_copy['tweet_id'] = t_api_df_copy['tweet_id'].astype(str)
```

```python
tweet_arch_copy['tweet_id'] = tweet_arch_copy['tweet_id'].astype(str)
```

```python
image_pred_copy['tweet_id'] = image_pred_copy['tweet_id'].astype(str)
```

```python
tweet_arch_copy= pd.merge(left=tweet_arch_copy,
                                 right=t_api_df_copy, left_on='tweet_id', right_on='tweet_id', how='inner')
```

```python
tweet_arch_copy = tweet_arch_copy.merge(image_pred_copy,on='tweet_id',how='inner')
```

### Issue #1: Some dog names are None and not NaN

#### If the dog name equal None assign it to NaN

#### Code

```python
tweet_arch_copy['name'] = tweet_arch_copy['name'].replace('None' , pd.np.nan)
```

```python
tweet_arch_copy['name'].value_counts()
```

### Issue #2:Some dog names doesn't have meaning (less than three characters)

#### If it lower than characters assign it to NaN

#### Code

```python
tweet_arch_copy['name'] = tweet_arch_copy['name'].where(tweet_arch_copy['name'].str.len() >= 3, pd.np.nan)
```

#### Test

```python
tweet_arch_copy['name'].value_counts()
```

### Issue #5:Dogs rating should be out of 10, and maximum rating should be 10

```python
tweet_arch_copy['rating_numerator'] = tweet_arch_copy['rating_numerator'].where(tweet_arch_copy['rating_numerator'] <= 10, 10)
tweet_arch_copy['rating_denominator'] = tweet_arch_copy['rating_denominator'].where(tweet_arch_copy['rating_denominator'] <= 10, 10)
```
